```python
import squares
import librosa
import matplotlib.pyplot as plt
import IPython.display as ipd
import numpy as np
import scipy
```

```python
x1, sr = librosa.load('recordings/widmung/2.mp3')
x2, sr = librosa.load('recordings/widmung/3.m4a')
```

```python
rms1 = librosa.feature.rms(x1)
rms2 = librosa.feature.rms(x2)
```

```python
plt.plot(np.arange(len(rms1[0])), rms1[0])
```

```python
def moving_avg(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
rms1_smooth = moving_avg(rms1[0], 400)
plt.plot(np.arange(len(rms1_smooth)), rms1_smooth)
```

```python
rms2_smooth = moving_avg(rms2[0], 500)
plt.plot(np.arange(len(rms2_smooth)), rms2_smooth)
```

```python
rms12 = librosa.feature.rms(x1, hop_length=22050)
rms12_smooth = moving_avg(rms12[0], 3)
plt.plot(np.arange(len(rms12[0])), rms12[0])
plt.plot(np.arange(len(rms12_smooth)), rms12_smooth)
p, phrases = get_phrases(rms12_smooth, 1)
```

```python
def get_phrases(t2, period, original=False):
    xt2 = np.arange(len(t2))

    xt2_fitted, xt2_phrases = squares.segls(len(t2), xt2, t2, 3)    
    return xt2_fitted, [(a[0] * period, a[1] * period) for a in xt2_phrases]
```

```python
def create_boundaries(arr, song_length):
    phrases = []
    last = 0
    for elem in arr:
        phrases.append((last, elem))
        last = elem + 1
    phrases.append((last, song_length))
    return phrases
```

```python
import datetime 
def convert(n): 
    return str(datetime.timedelta(seconds = n)) 
def print_boundaries(arr):
    for start, end in arr:
        a = convert(int(start))
        b = convert(int(end))
        print(a + " , " + b)
```

```python
print_boundaries(b)
```

```python
def find_peaks(x1):
    rms12 = librosa.feature.rms(x1, hop_length=22050)
    rms12_smooth = moving_avg(rms12[0], 10)
    rms12_maxes = scipy.signal.find_peaks(rms12_smooth)
    return rms12_smooth, rms12_maxes[0]
```

```python
etude1, sr = librosa.load('recordings/scriabin_etude1/1.mp3')
etude2, sr = librosa.load('recordings/scriabin_etude1/2.mp3')
```

```python
tcy, tcb = find_peaks(etude1)
hcy, hcb = find_peaks(etude2)
```

```python
plt.plot(np.arange(len(tcy)), tcy)
plt.plot(tcb, [tcy[a] for a in tcb], 'ro')
```

```python
plt.plot(np.arange(len(hcy)), hcy)
plt.plot(hcb, [hcy[a] for a in hcb], 'ro')
```
